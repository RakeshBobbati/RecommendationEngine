```python
df_eval['Eventtime_timestamp'] = pd.to_datetime(df_eval['Eventtime'],unit='ms')

class MakingInputFormat:

    def __init__(self,Wallet_Balance_name = "Wallet_Balance.csv"
                     ,df_train_name = 'train_data_4_except_first_game.csv'
                     ,demographic_data_name = "Demographic.csv" ):
        
        curr_dir = Path.cwd()
        self.df_WBalance = pd.read_csv(Path.joinpath(curr_dir,Wallet_Balance_name))
        self.df_train = pd.read_csv(Path.joinpath(curr_dir,df_train_name))
        self.df_demographic = pd.read_csv(Path.joinpath(curr_dir,demographic_data_name))

    def __making_input_data__(self,row):

        df_user_wallet = self.df_WBalance[self.df_WBalance.Uid ==  row.Uid]
        df_user_wallet["Eventtime_timestamp"] = row.Eventtime_timestamp
        df_user_wallet[' Timestamp'] = pd.to_datetime(df_user_wallet[' Timestamp'])
        df_user_wallet['time_diff']=df_user_wallet['Eventtime_timestamp']-df_eval_Wallet[' Timestamp']
        df_user_wallet=df_user_wallet[df_user_wallet['time_diff']>'0 days 00:00:00.00000']
        df_user_wallet=df_user_wallet.sort_values(by=['Uid','time_diff'])
        df_user_wallet['row_num_wallet_balance'] = df_user_wallet.groupby(['Uid']).cumcount()+1
        df_user_wallet["game_time_min"] = df_user_wallet[df_user_wallet.row_num_wallet_balance == 1]['time_diff'].dt.total_seconds().div(60)
        df_user_wallet["Wallet_Balance"] = df_user_wallet[df_user_wallet.row_num_wallet_balance == 2][' Wallet_Balance']


        df_train = self.df_train[self.df_train['Uid'] == row.Uid]

        if len(df_train)>0 and len(df_user_wallet)>0:
            try:
                wallet_balance = df_user_wallet[df_user_wallet.row_num_wallet_balance == 2][' Wallet_Balance'].values[0]
            except:
                wallet_balance = np.nan
            try:
                game_time_min = df_user_wallet[df_user_wallet.row_num_wallet_balance == 1]['time_diff'].dt.total_seconds().div(60).values[0]
            except:
                game_time_min = np.nan

            df_user_wallet = df_user_wallet[['Uid','Eventtime_timestamp',"Wallet_Balance","game_time_min"]].iloc[:1,:]
            df_train = pd.concat([df_train,df_user_wallet],axis = 0).reset_index().drop(columns= 'index')
            df_train["user_type"] = "previous"
            df_train.loc[df_train.index[-1], 'row_num'] = df_train.loc[df_train.index[-2], 'row_num']+1
            df_train.loc[df_train.index[-1], 'recent_entry_fee'] = df_train.loc[df_train.index[-2], 'Entry_Fee']
            df_train.loc[df_train.index[-1], 'recent_tournament'] = df_train.loc[df_train.index[-2], 'Tournament_Type']
            df_train.loc[df_train.index[-1], 'recent_num_players'] = df_train.loc[df_train.index[-2], 'Num_Players']
            df_train.loc[df_train.index[-1], 'cumulative_winning_sum'] = df_train.loc[df_train.index[-2], 'cumulative_winning_sum'] + df_train.loc[df_train.index[-2], 'Winnings']
            df_train.loc[df_train.index[-1], "Wallet_Balance_x"] = wallet_balance
            df_train.loc[df_train.index[-1], "game_time_min_x"] = game_time_min
            # print(df_train.loc[df_train.index[-1], "Wallet_Balance"])
            df_train.drop(columns = ['Wallet_Balance',"game_time_min"],inplace =True)
            # print(df_user_wallet[['Uid','Eventtime_timestamp',"Wallet_Balance","game_time_min"]].iloc[:1,:])
            
            
            for item in ["A","B","C","D"]:
                if item == df_train.loc[df_train.index[-2], 'Tournament_Type']:
                    df_train.loc[df_train.index[-1], item +'_winning_cumsum'] = df_train.loc[df_train.index[-2], item +'_winning_cumsum'] + df_train.loc[df_train.index[-2],'Winnings']
                    df_train.loc[df_train.index[-1], item +'_cumulative_count'] = df_train[df_train.Tournament_Type == item][item +'_cumulative_count'].max()+1
                    
                    if df_train.loc[df_train.index[-2], 'Win_Loss'] == 'W':
                        df_train.loc[df_train.index[-1], item +'_wins_entry_fee_sum'] = df_train.loc[df_train.index[-2], item +'_wins_entry_fee_sum'] + df_train.loc[df_train.index[-2],'Entry_Fee']
                        df_train.loc[df_train.index[-1], item +'_loss_entry_fee_sum'] = df_train[df_train.Tournament_Type == item][item +'_loss_entry_fee_sum'].max()
                        df_train.loc[df_train.index[-1], item +'_wins_previous_count'] = df_train[df_train.Tournament_Type == item][item +'_wins_previous_count'].max()+1
                    else:
                        df_train.loc[df_train.index[-1], item +'_loss_entry_fee_sum'] = df_train.loc[df_train.index[-2], item +'_loss_entry_fee_sum'] + df_train.loc[df_train.index[-2], 'Entry_Fee']
                        df_train.loc[df_train.index[-1], item +'_wins_entry_fee_sum'] = df_train[df_train.Tournament_Type == item][item +'_wins_entry_fee_sum'].max()
                        df_train.loc[df_train.index[-1], item +'_loss_previous_count'] = df_train[df_train.Tournament_Type == item][item +'_loss_previous_count'].max()+1

                else:
                    df_train.loc[df_train.index[-1], item +'_winning_cumsum'] = df_train[df_train.Tournament_Type == item][item +'_winning_cumsum'].max()
                    df_train.loc[df_train.index[-1], item +'_cumulative_count'] = df_train[df_train.Tournament_Type == item][item +'_cumulative_count'].max()
                    df_train.loc[df_train.index[-1], item +'_wins_entry_fee_sum'] = df_train[df_train.Tournament_Type == item][item +'_wins_entry_fee_sum'].max()
                    df_train.loc[df_train.index[-1], item +'_loss_entry_fee_sum'] = df_train[df_train.Tournament_Type == item][item +'_loss_entry_fee_sum'].max()

        else:
            Dict = {}
            for item in [list(df_train.columns) + ["user_type"]][0]:
                Dict[item] = [np.nan]
            df_train = pd.DataFrame.from_dict(Dict)
            df_train.user_type = 'new'
            df_train.Uid = row.Uid
            df_train.Eventtime_timestamp = row.Eventtime_timestamp
            
        return df_train.iloc[-1:,:].iloc[0]
        
making_input = MakingInputFormat()
```

```python
import pickle
from sklearn.cluster import KMeans

class LoadCatBoostModel:
    def __init__(self,model_path="./Tournament_classifier_val_acc_0.97_first_game_removed"):
        self.clf_catboost = CatBoostClassifier()
        try:
            self.clf_catboost.load_model(model_path)
        except Exception as e:
            raise e
    def __split_array__(self,arr):
        tournament = []
        entry_fee = []
        for item in arr:
            split = item.split("_")
            tournament.append(split[0])
            entry_fee.append(split[1])
        return tournament+entry_fee
            
    # Returns a pandas series of array eg:['A_25.0', 'A_50.0', 'C_25.0']
    def cat_predict(self,row):
        class_prob = np.argsort(self.clf_catboost.predict_proba(row))[::-1][:3]
        return pd.Series(self.__split_array__(self.clf_catboost.classes_[class_prob]))

class LoadKmeans:

    def __init__(self,model_path = "./clustering_model.pkl"
                ,city_tier_path = "./city_tier.csv"):

        self.kmeans = pickle.load(open("clustering_model.pkl", 'rb'))

        self.city_tier = pd.read_csv(city_tier_path)

        self.clusters = pd.read_csv("./clustering_classes_map.csv")

        self.encoding_connection_type=pd.read_pickle('./encoding_connection_type.pkl')
        self.encoding_cpu_type=pd.read_pickle('./encoding_Cpu_Type.pkl')
        self.encoding_network_type=pd.read_pickle('./encoding_Network_Type.pkl')
        self.encoding_isp=pd.read_pickle('./encoding_ISP.pkl')
        
        self.encoding_country_subdivision=pd.read_pickle('./encoding_Country_Subdivision.pkl')
        self.encoding_language=pd.read_pickle('./encoding_Language.pkl')
        self.encoding_device_type=pd.read_pickle('./encoding_Device_Type.pkl')
        self.encoding_device_name=pd.read_pickle('./encoding_Device_Name.pkl')
        
        self.encoding_encoding_device_manufacturer=pd.read_pickle('./encoding_Device_Manufacturer.pkl')
        
        self.encoding_postal_code=pd.read_pickle('./encoding_Postal_Code.pkl')
        self.encoding_city_tier=pd.read_pickle('./encoding_city_tier.pkl')

    def __predict_city_tier__(self,row):
            try:
                result = self.city_tier[self.city_tier.City == row.City]['city_tier'].values[0]
            except:
                result = 'Tier_1'
            return result

    def cluster_predict(self,df):
        df['city_tier'] = df.apply(self.__predict_city_tier__,axis = 1)
        df=df.merge(self.encoding_connection_type,on='Connection_Type')
        df=df.merge(self.encoding_cpu_type,on='Cpu_Type')
        df=df.merge(self.encoding_network_type,on='Network_Type')
        df=df.merge(self.encoding_isp,on='ISP')
        
        df=df.merge(self.encoding_country_subdivision,on='Country_Subdivision')
        df=df.merge(self.encoding_language,on='Language')
        df=df.merge(self.encoding_device_type,on='Device_Type')
        df=df.merge(self.encoding_device_name,on='Device_Name')
        df=df.merge(self.encoding_encoding_device_manufacturer,on='Device_Manufacturer')
        
        df=df.merge(self.encoding_postal_code,on='Postal_Code')
        df=df.merge(self.encoding_city_tier,on='city_tier')

        df['label'] = self.kmeans.fit_predict(df[['Connection_Type_frequency_encoded', 'Cpu_Type_frequency_encoded',
                                                   'Network_Type_frequency_encoded', 'ISP_frequency_encoded',
                                                   'Country_Subdivision_frequency_encoded', 'Language_frequency_encoded',
                                                   'Device_Type_frequency_encoded',
                                                   'Device_Manufacturer_frequency_encoded',
                                                   'Device_Name_frequency_encoded',
                                                   'Postal_Code_frequency_encoded', 'city_tier_frequency_encoded']])

        df[['Tournament_recommendation_1'
           ,'Tournament_recommendation_2'
           ,'Tournament_recommendation_3'
           ,'EntryFee_recommendation_1'
           ,'EntryFee_recommendation_2'
           ,'EntryFee_recommendation_3']] = df[['label']].apply(self.__category_pred__,axis = 1)

        return df

    def __split_array__(self,arr):
        tournament = []
        entry_fee = []
        for item in arr:
            split = item.split("_")
            tournament.append(split[0])
            entry_fee.append(split[1])
        return tournament+entry_fee

    def __category_pred__(self,row):     
        return pd.Series(self.__split_array__(self.clusters[str(row.label)].tolist()))
        
        
        
    
```
